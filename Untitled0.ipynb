```python
import cv2
import numpy as np
from pyzbar import pyzbar
import easyocr
import re
import time
import os
from PIL import Image
import matplotlib.pyplot as plt
from google.colab import files
import subprocess

# 1. Fixed defect detection function
def reliable_defect_detection(image_path):
    try:
        # Verify model exists
        best_model = "/content/runs/train/label_defect_model/weights/best.pt"
        if not os.path.exists(best_model):
            raise FileNotFoundError(f"Model not found at {best_model}")

        # Create unique output directory
        timestamp = int(time.time())
        output_dir = f"/content/runs/detect/verify_{timestamp}"
        os.makedirs(output_dir, exist_ok=True)

        # Run detection using subprocess to capture errors
        command = [
            "python", "/content/yolov5/detect.py",
            "--weights", best_model,
            "--source", image_path,
            "--conf", "0.4",
            "--save-txt",
            "--save-conf",
            "--project", "/content/runs/detect",
            "--name", f"verify_{timestamp}",
            "--exist-ok"
        ]
        result = subprocess.run(command, capture_output=True, text=True)

        if result.returncode != 0:
            raise RuntimeError(f"Detection failed: {result.stderr}")

        # Process results
        result_files = os.listdir(output_dir)
        result_images = [f for f in result_files if f.endswith('.jpg')]
        defects = []
        result_img = None

        if result_images:
            result_path = os.path.join(output_dir, result_images[0])
            if os.path.exists(result_path):
                result_img = Image.open(result_path)

                # Parse detection results
                txt_files = [f for f in result_files if f.endswith('.txt')]
                if txt_files:
                    with open(os.path.join(output_dir, txt_files[0])) as f:
                        for line in f:
                            parts = line.strip().split()
                            if len(parts) >= 5:  # Ensure valid detection line
                                class_id = int(parts[0])
                                defects.append(["crumpled", "torn", "ink_smear", "misprinted"][class_id])

        return result_img, defects

    except Exception as e:
        print(f"‚ùå Defect detection error: {str(e)}")
        return None, []
```

```python
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import random
import cv2
import matplotlib.pyplot as plt
import os

# Faster base label generation
def generate_base_label(device_id, batch_id, rohs=True):
    img = Image.new('RGB', (800, 400), (255, 255, 255))
    draw = ImageDraw.Draw(img)

    try:
        # Use a font that is likely to exist in Colab environments
        font_path = "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf"
        if not os.path.exists(font_path):
             font_path = "/usr/share/fonts/truetype/dejavu/DejaVuSans.ttf"
             if not os.path.exists(font_path):
                 font = ImageFont.load_default()
             else:
                 font = ImageFont.truetype(font_path, 24)
        else:
            font = ImageFont.truetype(font_path, 24)

    except Exception as e:
        print(f"Could not load truetype font, using default: {e}")
        font = ImageFont.load_default()

    # Device information
    draw.text((50, 50), f"Device ID: {device_id}", fill="black", font=font)
    draw.text((50, 100), f"Batch ID: {batch_id}", fill="black", font=font)

    # Compliance status
    status = "RoHS Compliant ‚úÖ" if rohs else "Non-RoHS ‚ö†Ô∏è"
    color = "green" if rohs else "red"
    draw.text((50, 150), status, fill=color, font=font)

    # Barcode simulation
    for i in range(15):
        width = random.randint(10, 30)
        x = 60 + i * 50
        draw.rectangle([x, 200, x + width, 300], fill="black")

    # QR code simulation
    for i in range(5):
        for j in range(5):
            if random.random() > 0.4:
                draw.rectangle([450 + i*60, 50 + j*60, 450 + i*60 + 40, 50 + j*60 + 40], fill="black")

    return img

# Optimized defect functions
def add_crumple_effect(img):
    img = np.array(img)
    h, w = img.shape[:2]
    scale = 0.05
    dx = scale * np.random.randn(h, w).astype(np.float32)
    dy = scale * np.random.randn(h, w).astype(np.float32)
    map_x = (np.tile(np.arange(w), (h, 1)) + dx * 50).astype(np.float32) # Ensure float32 type
    map_y = (np.tile(np.arange(h), (w, 1)).T + dy * 50).astype(np.float32) # Ensure float32 type
    # Use the second form of remap: two single-channel maps of type CV_32FC1 (np.float32)
    distorted = cv2.remap(img, map_x, map_y, cv2.INTER_LINEAR)
    return Image.fromarray(distorted)


def add_tear_effect(img):
    draw = ImageDraw.Draw(img, 'RGBA')
    w, h = img.size
    y = random.randint(100, h-100)
    for x in range(0, w, 10):
        y += random.randint(-15, 15)
        draw.line([(x, y), (x+10, y)], fill=(100,100,100,200), width=3)
    return img

def add_ink_smear(img):
    draw = ImageDraw.Draw(img, 'RGBA')
    w, h = img.size
    start = (random.randint(100, w-100), random.randint(100, h-100))
    for _ in range(3):
        end = (start[0] + random.randint(-80,80), start[1] + random.randint(-80,80))
        draw.line([start, end], fill=(0,0,0,150), width=random.randint(10,20))
        start = end
    return img

def add_misprint_effect(img):
    # Random color shift
    r, g, b = img.split()
    r = r.point(lambda p: p * random.uniform(0.7, 1.3))
    g = g.point(lambda p: p * random.uniform(0.7, 1.3))
    b = b.point(lambda p: p * random.uniform(0.7, 1.3))
    img = Image.merge('RGB', (r, g, b))

    # Add missing text
    if random.random() > 0.3:
        draw = ImageDraw.Draw(img)
        draw.rectangle([50, 50, 300, 180], fill="white")

    return img

# Generate dataset
def generate_dataset_samples(num_images=200, split_ratio=0.8):
    class_map = {0: "crumpled", 1: "torn", 2: "ink_smear", 3: "misprinted"}

    for i in range(num_images):
        # Create base label
        device_id = f"DVC{random.randint(10000, 99999)}"
        batch_id = f"BATCH-{random.randint(1000, 9999)}"
        base_img = generate_base_label(device_id, batch_id, rohs=random.random() > 0.3)

        # Determine split
        split = 'train' if i < num_images * split_ratio else 'val'

        # Apply defects or keep good
        if random.random() < 0.8:  # 80% defective
            defect_type = random.randint(0, 3)
            defect_name = class_map[defect_type]

            if defect_type == 0: img = add_crumple_effect(base_img)
            elif defect_type == 1: img = add_tear_effect(base_img)
            elif defect_type == 2: img = add_ink_smear(base_img)
            else: img = add_misprint_effect(base_img)

            # Save image
            img_path = f"/content/dataset/{split}/images/label_{i}_{defect_name}.jpg"
            img.save(img_path)

            # Create annotation
            label_path = f"/content/dataset/{split}/labels/label_{i}_{defect_name}.txt"
            # The bounding box coordinates (0.5 0.5 0.9 0.9) seem generic.
            # You might want to calculate more accurate bounding boxes for the defect area
            # or the label as a whole, depending on your YOLO training objective.
            # For this fix, we keep the generic box.
            with open(label_path, 'w') as f:
                f.write(f"{defect_type} 0.5 0.5 0.9 0.9\n")
        else:  # 20% good
            img_path = f"/content/dataset/{split}/images/label_{i}_good.jpg"
            base_img.save(img_path)

    print(f"Generated {num_images} samples")

# Generate dataset
generate_dataset_samples(200)

# Verify creation
print("\nDataset verification:")
print(f"Train images: {len(os.listdir('/content/dataset/train/images'))}")
print(f"Train labels: {len(os.listdir('/content/dataset/train/labels'))}")
print(f"Val images: {len(os.listdir('/content/dataset/val/images'))}")
print(f"Val labels: {len(os.listdir('/content/dataset/val/labels'))}")
```

```python
import os
import time
from PIL import Image
from IPython.display import display
from google.colab import files
import torch

def reliable_defect_detection(image_path, conf_thres=0.25, imgsz=640):
    """Robust defect detection with automatic device selection"""
    # Auto-select device (GPU if available, else CPU)
    device = "0" if torch.cuda.is_available() else "cpu"

    # Verify model exists
    best_model = "/content/runs/train/label_defect_model/weights/best.pt"
    if not os.path.exists(best_model):
        # Add a print statement to help debug if the file is truly missing
        model_weight_dir = os.path.dirname(best_model)
        print(f"Debug: Checking directory {model_weight_dir}")
        if os.path.exists(model_weight_dir):
            print(f"Debug: Contents of {model_weight_dir}: {os.listdir(model_weight_dir)}")
        else:
            print(f"Debug: Model directory {model_weight_dir} does not exist.")
        raise FileNotFoundError("‚ùå Trained model not found. Train first with train.py!")

    # Create unique output directory
    timestamp = int(time.time())
    output_dir = f"/content/runs/detect/verify_{timestamp}"
    os.makedirs(output_dir, exist_ok=True)

    # Run detection
    # Note: Using subprocess might be more robust for capturing errors
    # but for simplicity in Colab, the shell execution is often used.
    # If using subprocess, ensure commands and arguments are correctly formatted.
    !python /content/yolov5/detect.py \
      --weights {best_model} \
      --source "{image_path}" \
      --conf {conf_thres} \
      --imgsz {imgsz} \
      --device {device} \
      --augment \
      --save-txt \
      --save-conf \
      --project /content/runs/detect \
      --name verify_{timestamp} \
      --exist-ok

    # Process results
    detection_dir = f"/content/runs/detect/verify_{timestamp}"
    defects = []

    # Check for labels
    if os.path.exists(f"{detection_dir}/labels"):
        for label_file in os.listdir(f"{detection_dir}/labels"):
            if label_file.endswith(".txt"):
                with open(f"{detection_dir}/labels/{label_file}") as f:
                    for line in f:
                        class_id = int(line.split()[0])
                        defects.append(["crumpled", "torn", "ink_smear", "misprinted"][class_id])

    # Display results
    result_image_path = None
    for ext in ['.jpg', '.png', '.jpeg']:
        possible_path = f"{detection_dir}/{os.path.splitext(os.path.basename(image_path))[0]}{ext}"
        if os.path.exists(possible_path):
            result_image_path = possible_path
            break

    if result_image_path:
        result_img = Image.open(result_image_path)
        display(result_img)
    else:
        print(f"‚ö†Ô∏è No result image found in {detection_dir}")
        # Try to find any image file in the directory
        for file in os.listdir(detection_dir):
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                result_img = Image.open(f"{detection_dir}/{file}")
                display(result_img)
                break
        else:
            print("No images generated in output directory")

    return list(set(defects))  # Return unique defects

# Upload and test with automatic filename cleaning
print("üîº Please upload your image:")
uploaded = files.upload()
if uploaded:
    original_name = list(uploaded.keys())[0]
    clean_name = original_name.replace(" ", "_").replace("(", "").replace(")", "")
    if original_name != clean_name:
        os.rename(original_name, clean_name)

    print(f"\nüîç Analyzing {clean_name}...")
    print(f"‚öôÔ∏è Using {'GPU' if torch.cuda.is_available() else 'CPU'} for detection")
    # Call the function
    defects = reliable_defect_detection(clean_name)
    print(f"\n‚úÖ Detected defects: {defects if defects else 'None'}")
else:
    print("\n‚ùå No file uploaded!")



```
